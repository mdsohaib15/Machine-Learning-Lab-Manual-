```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
df1 = pd.read_csv('student-mat.csv',sep=';', encoding='latin-1')
df1.head()

```

```python
pd.set_option('display.max_columns', None)
df2 = pd.read_csv('student-por.csv', sep=';', encoding='latin-1')
df2.head()
```

```python
df1.shape, df2.shape
```

```python
#  merge both dataset
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
df.shape
```

```python
#  Data preprocessing and cleaning
df.isnull().sum()
```

No null value in both dataset

```python
# scaling
df_numeric = df.select_dtypes(include=[np.number])
df_numeric.head()

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_scaled = pd.DataFrame(scaler.fit_transform(df_numeric), columns=df_numeric.columns)
df_scaled.head()
```

```python
# visualize scaled data, my task is predicts a student will pass or fail based on study hours,attendence and assignment

plt.figure(figsize=(10,6))
sns.histplot(df_scaled['G3'], bins=30, kde=True)
plt.title('Distribution of Scaled Final Grades (G3)')
plt.xlabel('Scaled G3')
```
